# Web kazıma ile Bist hisse kodlarının alınması #

## 1) Kütüphanelerin indirilmesi

```python
import requests                     #bir web sayfasına istek göndermeye yarar, web sitesine tıklamak gibi. Biz bundan belirlediğimiz web sitesine istek göndermek için yararlanacağız.
from bs4 import BeautifulSoup       
import re                           #Pythonda Düzenli İfadeler (Regular Expressions) için oluşturulmuş bir modül. İstediğimiz bir metin içerisindeki belirli ifadeleri almak için kullanacağız.
```

## 2) Requests ile web sitemize istek gönderelim ve bağlanalım.

```python
URL = 'https://www.kap.org.tr/tr/bist-sirketler'
page = requests.get(URL)        # URL yi page değişkenine atıyorum.
```

## 3) Html parser ile page değişkenindeki içeriği ayrıştırma

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print (soup)  # Soup değişkenimize bir göz attığımızda, şu ana kadar yaptığımız işlemin sayfanın tüm html kodunu indirdiğini görebiliriz
```

## 4) Sadece almak istedimiz verilerin olduğu bloğu ayrıştırma

```python
results = soup.findAll("div", {"class": "comp-cell _04 vtable"})

#Almak istediğimiz blok “comp-cell _04” adlı sınıfın içerisinde yer alıyor. 
#Bunu ilgili siteye gelip browserda “incele” dedik ve sağ alanda ya da aşağıda çıkan html kodlarına bakarak bulduk.
```

```python
print(results)   # gelen veriler hala karışık
```

```python
### Kodların istediğimiz şekilde gelmesi için;

# a- Sonuçları stringe çeviriyoruz. Bu texti işleyebilmemiz için önemli.

# b- Boşlukları kaldırıyoruz.

# c- Regular Expressions ile ">" ve “</a>” arasında kalan hisse kodlarını ayıklıyoruz. Bunu words değişkenine atıyoruz.

# d- Boş bir liste oluşturuyoruz. Bunu daha sonra kullanacağız.

# e- oluşturduğumuz for döngüsü “>”, “</a” ve “,” karakterlerini atmamıza yarıyor. Her bir elemanı tarayarak işliyor.

# f- Bazı şirketlerin hisse kodları A1CAP, ACP şeklinde yan yana. Bunu ayıklayabilmek için birden fazla kod içeren satırlar (elemanlar) için listeye alt alta ekliyoruz.
```

```python
import sys
results = str(results)
results = results.replace('  ' ,'')
words = re.findall(r'>\S+</a>',results)
list1=[]
for i in words:  
    i = i.replace('>','')
    i = i.replace('</a','')
    i = i.split(',')
    if len(i)>1:
        for x in range(len(i)):
            list1.append(i[x])
    else:
        list1.append(i[0])
```

```python
print(list1)
```

```python
list2 = list1
 
for i in list2:
    print(i)         # tüm verileri alt alta sıraladım.
```
